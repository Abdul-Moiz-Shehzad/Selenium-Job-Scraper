```python
import pandas as pd
```

```python
df=pd.read_csv('Data Science Jobs 2024.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
print("before",len(df))
df=df.drop_duplicates()
print("after",len(df))
```

```python
df=df.loc[df["Salary Estimate"]!='-1']
len(df)
```

* Salary

```python
df["Salary Estimate"].unique()
```

```python
df["Employer est."]=df["Salary Estimate"].apply(lambda x: 1 if x.find('Employer est.')!=-1 else 0)
```

```python
df["Per hour"]=df["Salary Estimate"].apply(lambda x: 1 if x.find('Per Hour')!=-1 else 0)
```

```python
EstRemoved=df["Salary Estimate"].apply(lambda x: x.split(' (')[0])
```

```python
removed_dk=EstRemoved.apply(lambda x: x.replace('$','').replace('K',''))
df["Min Salary"]=removed_dk.apply(lambda x: float(x.split(' - ')[0]))
removed_dk
```

```python
removed_perhour=removed_dk.apply(lambda x: x.replace(' Per Hour',''))
removed_perhour
```

```python
df["Max Salary"]=removed_perhour.apply(lambda x: float(x.split(' - ')[1]) if x.find(' - ')!=-1 else float(x))
```

```python
df["Avg Salary"]=(df["Min Salary"]+df["Max Salary"])/2
```

* Location/State field

```python
State=df["Location"].apply(lambda x: x.split(', ')[1] if x.find(', ')!=-1 else x)
```

Florida
Virgina

```python
FL_Fixed=State.apply(lambda x: "FL" if x.find('Florida')!=-1 else x)
Virginia_Fixed=FL_Fixed.apply(lambda x: "VA" if x.find("Virginia")!=-1 else x)
Texas_Fixed=Virginia_Fixed.apply(lambda x: "TX" if x.find("Texas")!=-1 else x)
Fulton=Texas_Fixed.apply(lambda x: "NY" if x.find("Fulton")!=-1 else x)
df["State"]=Fulton
```
